```python
import numpy as np
import matplotlib.pyplot as plt
```

Consider $$\sigma_C(x) = \sup_{z \in C} z^T x.$$

Show something about where the subgradient method goes with either constant step size or with a step size sequence that is not summable, but it is square summable.

Assumption: C is convex, bounded, closed and proper

suppose further $x \in R$, $C=\{z \mid 1 \leq z \leq 3\}$

Then $\sigma_C(x)=\begin{cases}
    3x, x \geq 0\\
    x, x \lt 0
\end{cases}$

so that $\partial \sigma_C(x)=\begin{cases}
    3, x \gt 0\\
    [1,3], x=0 \\
    1, x \lt 0
\end{cases}$

```python
def piecewise_function(x):
    if x < 0:
        return x
    else:
        return 3*x

x_arr = np.linspace(-10, 5, 400)
y_arr = np.vectorize(piecewise_function)(x_arr)

plt.plot(x_arr, y_arr)
plt.xlabel('x')
plt.ylabel('y')
# plt.title('Piecewise Function Plot')
# plt.grid(True)
plt.show()
```

```python
def sub_grad(x):
    if x < 0:
        return 1
    elif x==0:
        t=np.random.uniform(low=1, high=3)
        return t
    else:
        return 3
```

```python
x0=3
step_size=1e-2
x=x0

record_x = []
record_y = []
for i in range(1000):
    x -= step_size * sub_grad(x)
    y = piecewise_function(x)
    if i % 30 == 0:
        record_x.append(x)
        record_y.append(y)

plt.plot(x_arr, y_arr)

plt.scatter(record_x, record_y,s=4,marker="x",color="red")
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
